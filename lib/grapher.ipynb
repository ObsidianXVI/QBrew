```python
import csv
import numpy as np
import matplotlib.pyplot as plt

fileToAccess: str = "../data/sm_g1_v0.csv"
sampleRate: int = 60

timesteps:list[int] = []
reward: list[float] = []

with open(fileToAccess) as f:
    plots = csv.reader(f, delimiter=';')
    next(plots)
    for i, row in enumerate(plots):
        if i % sampleRate !=0:
            continue
        timesteps.append(int(row[0]))
        reward.append(float(row[7]))
```

```python
plotName: str = "SoftMax"
plt.scatter(x=timesteps, y=reward)
plt.xlabel("timestep")
plt.ylabel("reward")
plt.title(plotName)

z = np.polyfit(timesteps, reward, 20)
p = np.poly1d(z)
plt.grid(linestyle="--", linewidth=0.5)
plt.plot(timesteps, p(timesteps), "r")
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

testToAccess: str = "sm1"
fileToAccess: str = "../../data/archives/" + testToAccess + "/data.csv"
sampleStepSize: int = 200 # 60
startIndex: int = 0

timesteps:list[int] = []
reward: list[float] = []

with open(fileToAccess) as f:
    plots = csv.reader(f, delimiter=';')
    next(plots)
    for i, row in enumerate(plots):
        if i < startIndex: continue
        if i % sampleStepSize !=0: continue
        timesteps.append(int(row[0]))
        reward.append(float(row[len(row)-1]))
```

```python
plt.scatter(x=timesteps, y=reward)
plt.xlabel("timestep")
plt.ylabel("reward")
plt.title(testToAccess)

z = np.polyfit(timesteps, reward, 4)
p = np.poly1d(z)
plt.grid(linestyle="--", linewidth=0.5)
plt.plot(timesteps, p(timesteps), "r")
```
